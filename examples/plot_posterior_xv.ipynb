```python
import pandas as pd
import os
%matplotlib inline
```

### In the following cell, you can choose the Patient_id and stimulation_type

```python
Patient_id = 'Patient2' # 'Patient2'
stimulation_type = 'SEEG' # 'TI'
```

```python
data_save_folder= '../../data/'
#os.makedirs(f'{data_save_folder}', exist_ok=True)
filename = f'{data_save_folder}/Patient2_{stimulation_type}_stimulation.xlsx'
df = pd.read_excel(filename)
```

```python
import numpy as np
```

```python
Roi_list = df.keys().tolist()
```

```python
x0_plot = df.values
```

```python
np.shape(x0_plot)
```

```python
import matplotlib.pyplot as plt
figsize = (8, 4.5)

fig, ax = plt.subplots(figsize=figsize)

regions_range = range(len(Roi_list))
vp = ax.violinplot(x0_plot, positions=range(len(Roi_list)), showmedians=True, widths = 0.8,vert=False)
ax.set_xlim([-0.1, 1.1])
#
#ax.axhline(y=-2.05, color='r', linestyle='--', lw=0.5)
#ax.axhline(y=-3.9, color='r', linestyle='--', lw=0.5)
ax.set_yticks(range(len(Roi_list)))
#ax.set_yticklabels(Roi_list, rotation='vertical',fontsize = 16);
ax.set_yticklabels(Roi_list, fontsize = 20)

#ax.set_xticklabels(ax.get_yticklabels(),fontsize = 16);

for pc in vp['bodies']:
    pc.set_facecolor('darkgreen')

#ax.axvline(x=0.6, color='r', linestyle='--', lw=0.6)
ax.invert_yaxis()
ax.tick_params(axis='both',labelsize = 20)
```
